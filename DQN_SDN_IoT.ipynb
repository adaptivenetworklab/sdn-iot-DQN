```python
def determine_action(row):
    # SLA camera
    if row["delay_ms"] > 100:
        return 0  # Prioritaskan Camera

    # SLA heart rate
    elif row["delay_ms"] > 50:
        return 1  # Prioritaskan Heart Rate

    # Jika delay aman, bantu suhu
    elif row["bw_suhu"] < row["bw_camera"] and row["bw_suhu"] < row["bw_heartrate"]:
        return 2  # Prioritaskan Suhu

    else:
        return 3  # Seimbang
```

```python
def simulate_next_state(row):
    bw_c, bw_h, bw_s = row["bw_camera"], row["bw_heartrate"], row["bw_suhu"]
    delay = row["delay_ms"]
    a = row["action"]

    if a == 0:  # Camera
        bw_c *= 1.10
        bw_h *= 0.95
        bw_s *= 0.95
        delay *= 0.90

    elif a == 1:  # Heart rate
        bw_h *= 1.10
        bw_c *= 0.95
        bw_s *= 0.95
        delay *= 0.93

    elif a == 2:  # Suhu
        bw_s *= 1.15
        bw_c *= 0.95
        bw_h *= 0.95

    else:  # Seimbang
        pass

    return pd.Series([bw_c, bw_h, bw_s, delay])
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque
```

```python
state_cols = ["bw_camera", "bw_heartrate", "bw_suhu", "delay_ms"]
next_state_cols = ["next_bw_camera", "next_bw_heartrate", "next_bw_suhu", "next_delay"]

states = dqn_dataset[state_cols].values
actions = dqn_dataset["action"].values
rewards = dqn_dataset["reward"].values
next_states = dqn_dataset[next_state_cols].values
dones = dqn_dataset["done"].values
```

```python
class ReplayBuffer:
    def __init__(self, capacity=100000):
        self.buffer = deque(maxlen=capacity)

    def push(self, s, a, r, s2, d):
        self.buffer.append((s, a, r, s2, d))

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        s, a, r, s2, d = zip(*batch)
        return (
            torch.FloatTensor(s),
            torch.LongTensor(a),
            torch.FloatTensor(r),
            torch.FloatTensor(s2),
            torch.FloatTensor(d)
        )

    def __len__(self):
        return len(self.buffer)
```

```python
buffer = ReplayBuffer()

for i in range(len(states)):
    buffer.push(states[i], actions[i], rewards[i], next_states[i], dones[i])

print("Total samples in buffer:", len(buffer))
```

```python
class DQN(nn.Module):
    def __init__(self, state_dim, action_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, action_dim)
        )

    def forward(self, x):
        return self.net(x)
```

```python
state_dim = states.shape[1]   # 4
action_dim = 4               # camera, HR, suhu, seimbang

policy_net = DQN(state_dim, action_dim)
target_net = DQN(state_dim, action_dim)
target_net.load_state_dict(policy_net.state_dict())

optimizer = optim.Adam(policy_net.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()

gamma = 0.99
batch_size = 64
epochs = 1500
target_update = 10
```

```python
for epoch in range(epochs):
    total_loss = 0
    steps = len(buffer) // batch_size

    for _ in range(steps):
        s, a, r, s2, d = buffer.sample(batch_size)

        # Q(s, a)
        q_values = policy_net(s).gather(1, a.unsqueeze(1)).squeeze()

        # max Q(s', a')
        next_q = target_net(s2).max(1)[0].detach()

        # Target Q
        target = r + gamma * next_q * (1 - d)

        loss = loss_fn(q_values, target)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if epoch % target_update == 0:
        target_net.load_state_dict(policy_net.state_dict())

    if epoch % 20 == 0:
        print(f"Epoch {epoch:03d} | Loss: {total_loss:.4f}")
```

```python
def select_action(state):
    with torch.no_grad():
        q = policy_net(torch.FloatTensor(state))
        return torch.argmax(q).item()
```
